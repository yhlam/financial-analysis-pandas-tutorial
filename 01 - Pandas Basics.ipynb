## Loading CSV file

Reading CSV file with pandas is simple. Pandas provides the read_csv function to read a CSV to a DataFrame. By default, it assumes that the fields are comma-separated.

We are going to try it with some Hong Kong weather data. The data we are using is provided by [Open Gov Project](http://opengov.jmsc.hku.hk/). We have included a copy in this repository. You can also download the up-to-date data from [here](http://data.jmsc.hku.hk/hongkong/observatory/daily_metob.csv).

```python
weather = pd.read_csv('data/hk_daily_weather.csv')
weather
```

read_csv also supports fetching data from the internet

```python
weather = pd.read_csv('http://data.jmsc.hku.hk/hongkong/observatory/daily_metob.csv')
weather.head()
```

Use the date column as index and parse it into datetime object

```python
weather = pd.read_csv('data/hk_daily_weather.csv', index_col='date', parse_dates=True)
weather.head()
```

Basic info of a DataFrame

```python
weather.info()
```

Statistic of a data frame

```python
weather.describe()
```

## Selecting columns from DataFrame

When you read a CSV, you get a kind of object called a DataFrame, which is made up of rows and columns.

Each column is a Series in pandas.

You get columns out of a DataFrame the same way you get elements out of a dictionary. 

```python
weather['meantemp']
```

You can also treat a column as attribute

```python
weather.meantemp
```

Select multiple columns by giving a list of column name

```python
weather[['maxtemp', 'meantemp', 'mintemp']]
```

## Visualization

You can use .plot() to plot a DataFrame or Series

```python
weather[['maxtemp', 'meantemp', 'mintemp']].plot()
```

```python
weather['meantemp'].plot()
```

Pandas can plot many types of chart, e.g. scatter plot. You just need to specifiy the kind parameter in the plot() function

```python
weather.plot(kind='scatter', y='pressure', x='meantemp')
```

Scatter plot matrix is useful for getting the overview of relationship among the data

```python
from pandas.tools.plotting import scatter_matrix

scatter_matrix(weather[['pressure', 'meantemp', 'relhum', 'rainfall']], figsize=(12, 12), diagonal='kde')
```

You can also plot the histogram and kernel density estimation

```python
weather['meantemp'].hist()  # histogram
```

```python
weather['meantemp'].plot(kind='kde')  # kernel density estimation
```

Combining several plots together into a single one is simple

```python
weather['meantemp'].hist(label='hist', normed=True)
weather['meantemp'].plot(kind='kde', label='kde', style='r')
plt.legend()
```

Beside overlaying the plots, you can plot them side-by-side as well

```python
fig, axes = plt.subplots(nrows=2, ncols=1)
plt.setp(axes[0], xlim=(-10, 50))
weather['meantemp'].hist(ax=axes[0])
weather['meantemp'].plot(ax=axes[1], kind='kde', xlim=(-10, 50))
```

## Filter

Select subset of the data base on some criteria is very intuitive in pandas

```python
meantemp = weather['meantemp']
meantemp[meantemp > 30]
```

Wanna know how many item are there in a series? Calling count() will do it

```python
meantemp[meantemp > 30].count()
```

```python
meantemp[meantemp > 30].count() / meantemp.count()
```

You can combine criteria by & and |

```python
weather[(weather['meantemp'] > 30) & (weather['relhum'] > 80)]
```

```python
weather.query('relhum >= 95 and rainfall > 100')
```

## Grouping

You can group data by some key like SQL.

```python
weather.groupby('relhum')['meantemp'].mean()
```

groupby() takes a function to generate key as well

```python
hot_days_by_year = meantemp[meantemp > 30].groupby(lambda date: date.year, axis=0).count()
hot_days_by_year
```

Visualize the data by ploting bar chart

```python
hot_days_by_year.plot(kind='bar')
```

```python
meantemp[meantemp < 10].groupby(lambda date: date.year, axis=0).count().plot(kind='bar', rot=False)
```

```python
meantemp_by_year = meantemp.groupby(lambda date: date.year, axis=0).mean()
meantemp_by_year
```

```python
meantemp_by_year.plot()
```

```python
meantemp.groupby(lambda date: date.month, axis=0).mean().plot()
```

```python
meantemp.groupby(lambda date: (date.year, date.month), axis=0).mean().plot()
```

```python
meantemp.groupby(lambda date: date.year, axis=0).std().plot()
```

```python
meantemp.groupby(lambda date: (date.year, date.month), axis=0).std().plot()
```
